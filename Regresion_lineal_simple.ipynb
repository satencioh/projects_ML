```python
import pandas as pd
import matplotlib.pyplot as plt
import random
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
```

```python
data = pd.read_csv("data/salarios.csv")
```

```python
X = data.iloc[:,:-1].values
y = data.iloc[:, 1].values
```

```python
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
def graficar_simple(data, data_t, colum):
    viz_trainx = plt
    viz_trainx.scatter(data[colum], data_t, color = 'green')
    viz_trainx.plot(data[colum], regressor.predict(data), color = 'black')
    viz_trainx.title('Salario vs ' + colum)
    viz_trainx.xlabel(colum)
    viz_trainx.ylabel('Salario')
    return viz_trainx.show()
```

```python
graficar_simple(X_train, y_train, 'Aexperiencia')
```

```python
graficar_simple(X_test,y_test, 'Aexperiencia')
```

```python
paises = ['CO','BR','PE','MX']
```

```python
le = preprocessing.LabelEncoder()
paises_codificados = le.fit_transform(paises)
```

```python
paises_dataset = [random.choice(paises_codificados) for i in range(len(data))]
```

```python
data['pais'] = paises_dataset
```

```python
X = data.drop('Salario', axis=1)
y = data.iloc[:, 1].values
```

```python
def graficar_multiple(data, data_t, column_1, column_2):
    fig = plt.figure(dpi = 150)
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(data[column_1],data[column_2],data_t, c='b',marker='s')
    ax.scatter(data[column_1],data[column_2],regressor.predict(data),c='r',marker='o')
    ax.set_xlabel(column_1)
    ax.set_ylabel(column_2)
    return ax.set_zlabel('$Salario$')
```

```python
graficar_multiple(X_train, y_train, 'Aexperiencia', 'pais')
```

```python
graficar_multiple(X_test, y_test, 'Aexperiencia', 'pais')
```

```python
regressor.score(X_test, y_test)
```
